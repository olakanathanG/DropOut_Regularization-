```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/sonar_dataset.csv",header=None)
```

```python
df.head(57)
```

```python
df.info()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(60,axis=1),df[60],test_size=0.2)
```

```python
label=preprocessing.LabelEncoder()
df[60]=label.fit_transform(df[60])

df[60].head()
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation="sigmoid")
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) # Changed 'complile' to 'compile'
model.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```

```python
modeld=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation="relu"),
    keras.layers.Dense(30,activation="relu"),
    keras.layers.Dense(15,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")
])
modeld.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) # Changed 'complile' to 'compile'
modeld.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
modeld.evaluate(x_test,y_test)
```
